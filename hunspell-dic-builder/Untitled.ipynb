```python
import json
```

```python
word_freqs = json.load(open("data/mg/mg_wordlist_freq.json"))
mg_word_freq = "data/mg/mg_wordlist.xml"
mg_word_list = "data/mg/mg_wordlist.txt"
root_open = """<wordlist locale="mg" description="Malagasy" date="1381226409" version="42">"""
root_close = """</wordlist>"""
word_template = """<w f="{}" flags="">{}</w>"""

# print(word_template.format(200,"hello"))
# print(word_freqs)
```

```python
with open(mg_word_freq, 'w') as wfreq, open(mg_word_list, 'r') as wlist:
    wfreq.write(root_open+"\n")
    for line in wlist:
        wfreq_line = ""
        _line = line.lower().strip().replace('\n','')
        if _line in word_freqs:
            wfreq_line = word_template.format(word_freqs[_line],_line)
        else:
            wfreq_line = word_template.format(0,_line)
        wfreq.write(wfreq_line+"\n")
    wfreq.write(root_close+"\n")
```
